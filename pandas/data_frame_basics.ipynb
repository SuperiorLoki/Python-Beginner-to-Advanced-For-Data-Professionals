```python
#Creating a new column
#We have to use the quotes to specify age
#Lambda is a quick way to do functions and it goes through all the x values in release_year and does the function

df["age"] = df.release_year.apply(lambda x: 2023 - x)
df.head(4)
```

```python
#Can use two columns to create a new column too
#You don't do anything near the apply but you have to specify what column you are talking about
#axis=1 specifies that we are looking row by row and x is a full row and not column
df["profit"] = df.apply(lambda x: x['revenue'] - x['budget'], axis = 1)
df.head(4)
```

```python
df["units"] = df.unit.apply(lambda x: x)
df.head(4)
```

```python
df.sort_values(by="profit", ascending=False).head(5)
```
