**If we want to group by the cities to find each of its maximum and minimum temperature or windspeed, we need to use group by method**

```python
#g is an object that is type groupby that is created 
g = df.groupby("city")
g
```

```python
#There are essentially three dataframes because there are three cities. So, one per city
for city, data in g:
    print("city", city)
    print("data", data)
```

**You want to create three groups**. 
  
**1. Days when temperature was between 80 and 90**.  
**2. Days when temperature was between 50 and 60**.   
**3. Days when it was anything else**. 

```python
def grouper(df, index, col):
    if 80 <= df[col].loc[index] <= 90:
        return '80-90'
    elif 50 <= df[col].loc[index] <= 60:
        return '50-60'
    else:
        return 'other'
```

```python
#Essentially, the lambda goes through each row of the data frame. 
#We also give it 'temperature' to specify the column so we can use those parameters in the grouper function
g = df.groupby(lambda index: grouper(df, index, 'temperature'))
```

```python
#When we analyze the results we can tell that the keys are the 80-90, 50-60, other criteria
for key, data in g:
    print("key:", key)
    print("data:", data)
```
