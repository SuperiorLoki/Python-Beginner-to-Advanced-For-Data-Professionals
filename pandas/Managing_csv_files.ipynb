```python
#To set the not available and n.a. values to a NaN value that is computer friendly
df = pd.read_csv("stock_data.csv", header = 1, na_values = {
 'eps': ['not available'], 'people': ['n.a.'], 'revenue': [-1], 'price': ['n.a.']
})
df
```

```python
#you can also look for those values in all the columns
df = pd.read_csv("stock_data.csv", header = 1, na_values = ['not available', -1, 'n.a.'])
df
```

```python
df["pe"] = df["price"]/df["eps"]
```

**Managing Excel Files**

```python
df_movies = pd.read_excel("movies_db.xlsx", "movies")
df_movies.head(4)
```

```python
df_financials = pd.read_excel("movies_db.xlsx", "financials")
df_financials.head(5)
```

```python
#to convert the weird characters to what we want, data cleaning

def standardize_currency(curr):
    if curr == "$$" or curr == "Dollars":
        return "USD"
    return curr

df_financials = pd.read_excel("movies_db.xlsx", "financials", converters = {
 'currency': standardize_currency   
})
df_financials.head(5)
```

```python
#to combine two pages
#The on is the column that we want to have common and it uses that to match the two
df_merged = pd.merge(df_movies, df_financials, on="movie_id")
df_merged.head(4)
```

```python
#export to excel file
df_merged.to_excel("movies_merged.xlsx", sheet_name="merged", index=False)
```

```python
#Create your own data frame using dicts
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})
df_stocks
```

```python
df_weather = pd.DataFrame({
    'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
    'temperature': [32, 35, 28],
    'event': ['Rain', 'Sunny', 'Snow']
})
df_weather
```

```python
#to combine both into a single file
with pd.ExcelWriter("stocks_weather.xlsx") as writer:
    df_stocks.to_excel(writer, sheet_name = "stocks")
    df_weather.to_excel(writer, sheet_name = "weather")
```
