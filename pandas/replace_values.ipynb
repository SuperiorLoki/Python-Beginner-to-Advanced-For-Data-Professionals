```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data2.csv")
df
```

```python
#you can replace specific value(s) that is present with whatever you want to replace it with
import numpy as np

df.replace([-99999, -88888], np.nan)
```

```python
#You can set specific values in specific columns that you want replaced

df.replace({
    'temperature': -99999,
    'windspeed': [-88888, -99999],
    'event': 'no event'
}, np.nan)
```

```python
#Alternatively, you can do the opposite. You can set what value you want to see depending on what you are replacing

df.replace({
    -99999: np.nan,
    -88888: np.nan,
    'no event': 'Normal'
})
```
